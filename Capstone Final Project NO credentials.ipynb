```python
#loading further advanced libaries

!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
#getting the geocoordinates for centering the map

address = 'New York City, NY'

geolocator = Nominatim()
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
```

```python
#processing data with the goal of making the geocoordinates usable for Foursquare and mapping

##how to split "the_geom" into two columns: https://www.geeksforgeeks.org/python-pandas-split-strings-into-two-list-columns-using-str-split/
##OR (maybe better): https://chrisalbon.com/python/data_wrangling/pandas_split_lat_and_long_into_variables/

df = pd.read_csv('https://data.cityofnewyork.us/api/views/cqsj-cfgu/rows.csv?accessType=DOWNLOAD')
df['the_geom'] = df['the_geom'].str[7:-1]

df.head()
```

```python
# new data frame with split value columns 
new = df['the_geom'].str.split(" ", n = 1, expand = True) 
  
# making seperate first part column from new data frame 
df['lon']= new[0].astype('float')
  
# making seperate last part column from new data frame 
df['lat']= new[1].astype('float')

# df display 
df.head()
```

```python
#creating a new dataframe for working with Foursquare

df02 = df[['lat', 'lon', 'Borough', 'Street_Nam']].copy()

df02.head()
```

```python
#defining my Foursquare credentials

CLIENT_ID = 'REMOVED' # your Foursquare ID
CLIENT_SECRET = 'REMOVED' # your Foursquare Secret
VERSION = '20180605' # Foursquare API version

print('Your credentails:')
print('CLIENT_ID: ' + CLIENT_ID)
print('CLIENT_SECRET:' + CLIENT_SECRET)
```

```python
#loading further libaries

import json # library to handle JSON files
import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe
```

```python
#obtain number of ALL restaurants for surrounding of each pedestrian counter

LIMIT = 100

radius = 300

# ID for restaurants: 4d4b7105d754a06374d81259
categoryId = '4d4b7105d754a06374d81259'

rest_list=[]

for i in range(0,len(df02)):
    url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}&categoryId={}'.format(
        CLIENT_ID, 
        CLIENT_SECRET, 
        VERSION, 
        df02.loc[i, 'lat'], 
        df02.loc[i, 'lon'],
        radius, 
        LIMIT,
        categoryId
    )
    results = requests.get(url).json()
    resultsNORM = json_normalize(results)
    rest_list.append(
        resultsNORM['response.totalResults'])
    
df03 = pd.DataFrame(rest_list)

df03.reset_index(inplace=True)

df03a = df03.drop(['index'], axis=1)

df03a.columns = df03a.columns.astype(str)

df03a.rename(columns={"0": "restaurants"}, inplace=True)

df03a.head()
```

```python
#joining the number of all restaurants with the main dataframe

dfa = df.join(df03a)

dfa.head()
```

```python
#obtain number of ITALIAN restaurants for surrounding of each pedestrian counter


LIMIT = 100

radius = 300

# ID for Italian restaurants:
categoryId2 = '4bf58dd8d48988d110941735'

ital_rest_list=[]

for i in range(0,len(df02)):
    url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}&categoryId={}'.format(
        CLIENT_ID, 
        CLIENT_SECRET, 
        VERSION, 
        df02.loc[i, 'lat'], 
        df02.loc[i, 'lon'],
        radius, 
        LIMIT,
        categoryId2
    )
    results = requests.get(url).json()
    resultsNORM = json_normalize(results)
    ital_rest_list.append(
        resultsNORM['response.totalResults'])


df04 = pd.DataFrame(ital_rest_list)

df04.reset_index(inplace=True)

df04a = df04.drop(['index'], axis=1)

df04a.columns = df04a.columns.astype(str)

df04a.rename(columns={"0": "italianrestaurants"}, inplace=True)

df04a.head()
```

```python
#joining the number of Italian restaurants with the main dataframe

dfb = dfa.join(df04a)

dfb.head()

```

```python
#calculating the frequency metric per counter

#take mean of PM May, Saturday May and PM September and Saturyday September, for 2015 to 2017

dfb['frequency'] = dfb[['May15_PM','May15_MD','Sept15_PM','Sept15_MD','May16_PM','May16_MD','Sept16_PM','Sept16_MD','May17_PM','May17_MD','Sept17_PM','Sept17_MD']].mean(axis=1)

dfb.head()
```

```python
#applying Z transformation for cluster analysis

dfb['Zrestaurants']=(dfb['restaurants']-dfb['restaurants'].mean())/dfb['restaurants'].std()

dfb['Zitalianrestaurants']=(dfb['italianrestaurants']-dfb['italianrestaurants'].mean())/dfb['italianrestaurants'].std()

dfb['Zfrequency']=(dfb['frequency']-dfb['frequency'].mean())/dfb['frequency'].std()

dfb.head()

#df_zscore = (df - df.mean())/df.std()
```

```python
#import libary for kmeans cluster analysis

from sklearn.cluster import KMeans
```

```python
#define dataframe for cluster analysis

dfb2 = dfb[['Zrestaurants', 'Zitalianrestaurants', 'Zfrequency']].copy()

dfb2.head()
```

```python
# set number of clusters
kclusters = 3

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(dfb2)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
#create column for cluster label

dfb2['cluster'] = kmeans.labels_

dfb2.head()
```

```python
#join cluster labels with main dataframe

dfb3 = dfb2.drop(['Zrestaurants', 'Zitalianrestaurants', 'Zfrequency'], axis=1)

dfc = dfb.join(dfb3)

dfc.head()
```

```python
#report means of critical variables basically defining the clusters

dfc.groupby('cluster')['restaurants', 'italianrestaurants', 'frequency'].mean()
```

```python
#more extended description of clusters

dfc.groupby('cluster')['restaurants', 'italianrestaurants', 'frequency'].describe()
```

```python
#mapping all pedestrian counters and indicating cluster by color

map_clusters = folium.Map(location=[latitude, longitude], zoom_start=10)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lng, Loc, cluster in zip(dfc['lat'], dfc['lon'], dfc['Loc'], dfc['cluster']):
    label = folium.Popup(' location ' + str(Loc) + '; ' +' cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters

```
